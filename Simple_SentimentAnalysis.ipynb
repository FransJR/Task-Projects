## Pre-processing Teks

```python
#Ubah menjadi lowercase
for i in range(len(text_data)):
    text_data[i] = text_data[i].lower()
```

```python
text_data
```

```python
#penghilangan Hashtag
for i in range(len(text_data)):
    text_data[i] = text_data[i].replace("#snmptn2021","")
    text_data[i] = text_data[i].replace("#snmptn","")
    text_data[i] = text_data[i].replace("#snmangat","")
text_data
```

```python
#penghilangan tanda baca
for i in range(len(text_data)):
    text_data[i] = text_data[i].replace(".","")
    text_data[i] = text_data[i].replace("(","")
    text_data[i] = text_data[i].replace(":","")
    text_data[i] = text_data[i].replace(")","")
    text_data[i] = text_data[i].replace(",","")
    text_data[i] = text_data[i].replace("!","")
    text_data[i] = text_data[i].replace("?","")
text_data
```

```python
# Pembenaran Typo
# Dikarenakan data yang sedikit tidak digunakan iterasi pada keseluruhan teks data
text_data[6] = text_data[6].replace("gapapaa", "gapapa")
text_data[11] = text_data[11].replace("piliha n", "pilihan")
```

```python
# lematisasi
for i in range(len(text_data)):
    text_data[i] = text_data[i].replace(" engga "," tidak ")
    text_data[i] = text_data[i].replace(" gak "," tidak ")
    text_data[i] = text_data[i].replace("bukan","tidak")
    text_data[i] = text_data[i].replace("jangan","tidak")
    text_data[i] = text_data[i].replace("lolos"," ulus")
    text_data[i] = text_data[i].replace("thanks","terimakasih")
    text_data[i] = text_data[i].replace("gugur","gagal")
    text_data[i] = text_data[i].replace("  "," ")    
text_data 
```

```python
# pemisahan per kata agar iterasi lebih cepat
text_data_word = []
for i in range(len(text_data)):
    text_data_word.append(list(np.array(text_data[i].split(" "))))
    text_data_word[i] = " ".join(text_data_word[i]).split()
text_data_word
```

## Kategorisasi Data

```python
# Deklarasi kata kunci
kata_negasi= ["tidak"]
#kata bermakna negatif
kata_negatif = ["putus asa", "gagal", "lolos", "menyerah"]
#kata yang negatif yang tetap bermakna negatif dengan atau tanpa negasi
kata_negatif_abs = [ "semangat", "menangis"]
#kata bermakna positif
kata_positif = ["lulus", "masuk"]
#kata yang positif yang tetap bermakna positif dengan atau tanpa negasi
kata_positif_abs = ["selamat", "alhamdulillah", "terimakasih"]
predicted_is_positive = [0]*12
```

```python
# mengecek sentimen yang ada pada teks(positif atau negatif)
# sentimen dihitung menggunakan score yang dihitung dari kata kunci pada teks
"Iterasi pada seluruh kalimat"
for i in range(len(text_data_word)):
    score = 0
    "pembatasan penggunaan kata negasi sesuai jumlahnya"
    num_negasi = text_data_word[i].count(kata_negasi[0])
    negasi_use = 0
    "Iterasi pada seluruh kata pada array"
    for j in range(len(text_data_word[i])):
        "Pengecekan makna negatif"
        for kata in kata_negatif:
            if( kata == text_data_word[i][j] ):
                if ((kata_negasi[0] in text_data_word[i]) and (negasi_use < num_negasi)):
                    "Mengecek jarak kata kunci dengan kata negasi"
                    "kata negasi harus berposisi minimal 2 kata depan kata kunci"
                    if((text_data_word[i].index(kata_negasi[0]) - (j+1)) < 0):
                        #print(text_data_word[i].index(kata_negasi[0]) - j," ",i,"\n")
                        score += 1
                        negasi_use += 1
                    else:
                        score -=1
                else:
                    score -= 1
                    
        "Pengecekan makna negatif absolut"
        for kata in kata_negatif_abs:
            if( kata == text_data_word[i][j] ):
                score -= 1
                
        "Pengecekan makna positif"
        for kata in kata_positif:
            if( kata == text_data_word[i][j] ):
                if ((kata_negasi[0] in text_data_word[i]) and (negasi_use < num_negasi)):
                    "Mengecek jarak kata kunci dengan kata negasi"
                    if((text_data_word[i].index(kata_negasi[0]) - j) < 0):
                        #print((text_data_word[i].index(kata_negasi[0]) - j),"\n")
                        score -= 1
                        negasi_use += 1
                    else:
                        score += 1
                else:
                    score += 1
                    
        "Pengecekan makna positif absolut"
        for kata in kata_positif_abs:
            if( kata == text_data_word[i][j] ):
                score += 1
                
    "Pengkategorisasian sentimen berdasarkan score terhitung"
    if score > 0:
        predicted_is_positive[i] = 1
    else:
        predicted_is_positive[i] = 0
            
```
