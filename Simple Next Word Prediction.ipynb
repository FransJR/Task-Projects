```python
# Import Dependencies
import numpy as np
import pandas as pd
```

```python
"Melakukan prediksi satu kata selanjutnya dengan input 1 buah kata"
class prediksikata:
    # definisi parameter input kelas
    # input berupa array of string 
    def __init__(self, referensi):
        self.len = len(referensi)
        
        #penghilangan tanda baca
        for i in range(self.len):
            referensi[i] = referensi[i].replace(".","")
            referensi[i] = referensi[i].replace("(","")
            referensi[i] = referensi[i].replace(":","")
            referensi[i] = referensi[i].replace(")","")
            referensi[i] = referensi[i].replace(",","")
            referensi[i] = referensi[i].replace("!","")
            referensi[i] = referensi[i].replace("?","")
        
        #penggabungan data menjadi 1 string
        string_temp = referensi[0]
        for i in range(self.len-1):
            string_temp += " . "
            string_temp += referensi[i+1]
            
        #print(string_temp)
        
        #pisah string menjadi perkata
        self.ref_kata = []
        self.ref_kata.append(list(np.array(string_temp.split(" "))))
        self.ref_kata = " ".join(self.ref_kata[0]).split()
        
        #update panjang kata
        self.len = len(self.ref_kata)
        
    # prediksi dengan membedakan lowercase dan uppercase
    def predict(self, input_kata):
        self.next_word = []
        self.count = []
        
        #mencari kata dalam referensi 
        for i in range (self.len):
            if (self.ref_kata[i] == input_kata):
                if(i < self.len):
                    temp = self.ref_kata[i+1]
                    if temp not in self.next_word:
                        self.next_word.append(temp)
                        self.count.append(1)
                    else:
                        self.count[self.next_word.index(temp)] += 1
        
        #mencari kata dengan frekuensi terbesar, bila terdapat frekuensi sama dipilih yang pertama ditemukan
        if len(self.next_word) != 0:
            return(self.next_word[self.count.index(max(self.count))])
        else:
            return self.next_word

    
    # prediksi dengan tidak membedakan lowercase dan uppercase (Case Insensitive)
    def predict_ci(self, input_kata):
        self.next_word = []
        self.count = []
        
        #mencari kata dalam referensi 
        for i in range (self.len):
            if (self.ref_kata[i].lower() == input_kata.lower()):
                if(i < self.len):
                    temp = self.ref_kata[i+1]
                    if temp not in self.next_word:
                        self.next_word.append(temp)
                        self.count.append(1)
                    else:
                        self.count[self.next_word.index(temp)] += 1
        
        #mencari kata dengan frekuensi terbesar, bila terdapat frekuensi sama dipilih yang pertama ditemukan
        if len(self.next_word) != 0:
            return(self.next_word[self.count.index(max(self.count))])
        else:
            return self.next_word
    
    
    # prediksi dengan input beberapa kata dan mencocokan dari n kata dan bila tidak ada hanya prediksi dengan input kata terakhir
    def predict_kalimat(self , kalimat):
        "Pengolahan inpu kalimat"
        #penghilangan tanda baca
        self.kalimat = kalimat
        self.kalimat = self.kalimat.replace(".","")
        self.kalimat = self.kalimat.replace("(","")
        self.kalimat = self.kalimat.replace(":","")
        self.kalimat = self.kalimat.replace(")","")
        self.kalimat = self.kalimat.replace(",","")
        self.kalimat = self.kalimat.replace("!","")
        self.kalimat = self.kalimat.replace("?","")
        
        #pisah string menjadi perkata
        self.kata_kunci = []
        self.kata_kunci.append(list(np.array(self.kalimat.split(" "))))
        self.kata_kunci = " ".join(self.kata_kunci[0]).split()
        #print(self.kata_kunci)
        
        #pencocokan  kata kunci
        count_kunci = len(self.kata_kunci)
        self.next_word = []
        self.count = []
        for i in range (self.len):
            miss = False
            if (self.ref_kata[i].lower() == self.kata_kunci[0].lower()):
                #print(self.kata_kunci[0])
                for j in range(count_kunci-1):
                    if(self.ref_kata[i+j+1].lower() != self.kata_kunci[j+1].lower()):
                        miss = True
                    #print(miss)
                if not(miss) :
                    temp= self.ref_kata[i + count_kunci ]
                    if temp not in self.next_word:
                        self.next_word.append(temp)
                        self.count.append(1)
                    else:
                        self.count[self.next_word.index(temp)] += 1  
                    
        #mencari kata dengan frekuensi terbesar, bila terdapat frekuensi sama dipilih yang pertama ditemukan
        if len(self.next_word) != 0:
            return(self.next_word[self.count.index(max(self.count))])
        else:
            "jika tidak ketemu hanya prediksi dari input kata terakhir saja"
            return self.predict(self.kata_kunci[count_kunci-1])
        
    
    # prediksi dengan input beberapa kata dan mencocokan dari n kata, n-1 kata, hingga 1 kata
    def predict_kalimat_nkata(self , kalimat):
        "Pengolahan input kalimat"
        #penghilangan tanda baca
        self.kalimat = kalimat
        self.kalimat = self.kalimat.replace(".","")
        self.kalimat = self.kalimat.replace("(","")
        self.kalimat = self.kalimat.replace(":","")
        self.kalimat = self.kalimat.replace(")","")
        self.kalimat = self.kalimat.replace(",","")
        self.kalimat = self.kalimat.replace("!","")
        self.kalimat = self.kalimat.replace("?","")
        
        #pisah string menjadi perkata
        self.kata_kunci = []
        self.kata_kunci.append(list(np.array(self.kalimat.split(" "))))
        self.kata_kunci = " ".join(self.kata_kunci[0]).split()
        #print(self.kata_kunci)
        
        #pencocokan  kata kunci
        count_kunci = len(self.kata_kunci)
        self.next_word = []
        self.count = []
        for i in range (self.len):
            miss = False
            if (self.ref_kata[i].lower() == self.kata_kunci[0].lower()):
                #print(self.kata_kunci[0])
                for j in range(count_kunci-1):
                    if(self.ref_kata[i+j+1].lower() != self.kata_kunci[j+1].lower()):
                        miss = True
                    #print(miss)
                if not(miss) :
                    temp= self.ref_kata[i + count_kunci ]
                    if temp not in self.next_word:
                        self.next_word.append(temp)
                        self.count.append(1)
                    else:
                        self.count[self.next_word.index(temp)] += 1  
                    
        #mencari kata dengan frekuensi terbesar, bila terdapat frekuensi sama dipilih yang pertama ditemukan
        if len(self.next_word) != 0:
            return(self.next_word[self.count.index(max(self.count))])
        else:
            "jika tidak ketemu lanjut ke prediksi selanjutnya"
            if(count_kunci > 2):
                #print(" ".join(self.kata_kunci[1:]))
                return self.predict_kalimat_nkata(" ".join(self.kata_kunci[1:]))
            else:
                return self.predict(self.kata_kunci[count_kunci-1])
```
